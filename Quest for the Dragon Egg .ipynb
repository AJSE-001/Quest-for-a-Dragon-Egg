```python
# Room descriptions
descriptions = {
    "dragonville": "You are in Dragonville, a peaceful village where your adventure begins.",
    "tower": "You are in the wizard's tower. The wizard makes potions here.",
    "forest": "You are in the enchanted forest. The elves live here and guard their secrets.",
    "cave": "You are in the dark cave where the dragon egg is kept. Solve the riddle to obtain it."
}

# Paths between rooms
paths = {
    "dragonville": ["tower", "forest"],
    "tower": ["dragonville"],
    "forest": ["dragonville", "cave"],
    "cave": ["forest"]
}

# Initialize game state
current_loc = "dragonville"
has_potion = False
has_map = False

print("""
Welcome to the Quest for the Dragon Egg
=========================

Your adventure begins in Dragonvill.
""")

while True:
    # Print the description of the current location
    print(descriptions[current_loc])
    
    # Check for interactions based on the current location
    if current_loc == "tower":
        if not has_potion:
            print("The wizard offers you a potion. You take it.")
            has_potion = True
        else:
            print("You have already taken the potion.")
    
    elif current_loc == "forest":
        if has_potion and not has_map:
            print("The elves accept your potion and give you a map to the cave.")
            has_map = True
        elif not has_map:
            print("The elves will not give you the map without a potion.")
        else:
            print("You have the map. You can go to the cave now.")
    
    elif current_loc == "cave":
        if has_map:
            print("You are at the cave. Solve the riddle to obtain the dragon egg.")
            riddle_answer = input("What has keys but can't open locks? ").strip().lower()
            if riddle_answer == "piano":
                print("Correct! You have obtained the dragon egg!")
                break
            else:
                print("Incorrect! You must solve the riddle to obtain the egg.")
                current_loc = "forest"  # Go back to the forest to try again
                continue
        else:
            print("You need a map to enter the cave.")
            current_loc = "forest"  # Go back to the forest to get the map
            continue
    
    # Print available paths
    print(f"You can go to: {', '.join(paths[current_loc])}.")
    
    # Get user input for the next location
    next_loc = input("Where do you want to go next? ").strip().lower()
    
    # Check if the input is a valid path
    if next_loc in paths[current_loc]:
        current_loc = next_loc
    else:
        print("You can't go that way!")

# Final message when the player finds the dragon egg
print("Congratulations! You have completed your quest!")
```
